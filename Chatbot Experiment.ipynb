### Warning & Error Suppression

```python
# Suppress warning & error messages from clogging the output space
import warnings
warnings.filterwarnings("ignore")

from transformers.utils import logging
logging.set_verbosity_error()

import os
os.environ["TRANSFORMERS_NO_ADVISORY_WARNINGS"] = "true"
os.environ["TOKENIZERS_PARALLELISM"] = "false"

os.environ["DISABLE_TQDM"] = "1"
```

### Text-Generation AI model

```python
from transformers import pipeline

# Loading a free open-source text generation model for chat bot
# I chose Tiny Llama from Hugging Face
generator = pipeline("text-generation", model="TinyLlama/TinyLlama-1.1B-Chat-v1.0")
```

```python
# System prompt (made to stay consistent)
system_msg = {
        "role": "system",
        "content": "You are a helpful AI assistant."
    }

# Keeping conversation history in a list to maintain context
messages = [system_msg]

# Initial greeting
print("\nHi there. What questions do you have for me? \n" \
"Type 'exit' to quit. \n")

# Chat interface loop
while True:

    # Getting user input
    user_input = input("You: ")

    # Allowing exit (also added quit and q for convenience)
    if user_input.lower() in ["exit", "quit", "q"]:
        print("See you later!")
        break

    # Having user input appearing in the output for user experience
    print(f"\nYou: {user_input}")

    # Adding user input to chat history
    messages.append({
        "role": "user",
        "content": user_input
    })

    # Converting chat history to prompt
    prompt = generator.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    # Generating response from the model with specified parameters
    output = generator(
        prompt,
        max_new_tokens=120,     # length of response parameter
        temperature=0.7,        # creativity parameter
        do_sample=True,         
        top_p=0.9, 
        top_k=50, 
        pad_token_id=generator.tokenizer.eos_token_id
    )

    # Extracting the generated text from the output
    full_response = output[0]["generated_text"]

    # Extract only new assistant text
    assistant_response = full_response[len(prompt):].strip()

    # Displaying assistant response
    print(f"\nAI: {assistant_response}\n")

    # Adding assistant response to chat history
    messages.append({
        "role": "assistant",
        "content": assistant_response
    })
```
